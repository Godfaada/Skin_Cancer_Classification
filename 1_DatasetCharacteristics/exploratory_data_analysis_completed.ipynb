# Exploratory Data Analysis: Skin Cancer Detection

This notebook performs initial analysis on the HAM10000 skin lesion dataset including data inspection, visualization, and label distribution.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_DIR = '.'
CSV_PATH = os.path.join(DATA_DIR, 'HAM10000_metadata.csv')

meta = pd.read_csv(CSV_PATH)
meta.head()
```

## Class Distribution

```python
sns.countplot(data=meta, x='dx')
plt.title('Lesion Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

## Gender Distribution

```python
sns.countplot(data=meta, x='sex')
plt.title('Gender Distribution')
plt.show()
```

## Age Distribution

```python
meta['age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
